# Prebuilt word embeddings model  

# Import Modules


```python
import gensim.downloader as api
```

# Load the pre-trained model


```python
# ============================================================
# Load the pre-trained model
# ============================================================
print("Loading model... (this may take a few minutes)")
model = api.load("word2vec-google-news-300")
print("Model loaded!\n")
```

# Test the model 

## ANALOGY Test 

```python

# ============================================================
# 1. ANALOGY: king - man + woman = ?
#    Add and subtract word vectors
# ============================================================
print("=" * 50)
print("1. ANALOGY (king - man + woman)")
print("=" * 50)
result = model.most_similar(
    positive=["king", "woman"],
    negative=["man"],
    topn=5
)
for word, score in result:
    print(f"  {word:20} → {score:.4f}")
```

## MOST SIMILAR: find closest words to a given word

```python
print("\n" + "=" * 50)
print("2. MOST SIMILAR to 'doctor'")
print("=" * 50)
result = model.most_similar("doctor", topn=5)
for word, score in result:
    print(f"  {word:20} → {score:.4f}")
```

## SIMILARITY: how similar are two words? (0 to 1)

```python
print("\n" + "=" * 50)
print("3. SIMILARITY between two words")
print("=" * 50)
pairs = [
    ("cat", "dog"),        # very similar
    ("cat", "car"),        # not similar
    ("king", "queen"),     # related
    ("happy", "sad"),      # opposites
]
for w1, w2 in pairs:
    score = model.similarity(w1, w2)
    print(f"  {w1:10} ↔ {w2:10} → {score:.4f}")
```

## DOESNT MATCH: find the odd word out in a list

```python
print("\n" + "=" * 50)
print("4. DOESNT MATCH (odd word out)")
print("=" * 50)
groups = [
    ["breakfast", "lunch", "dinner", "car"],
    ["cat", "dog", "fish", "computer"],
    ["paris", "london", "berlin", "pizza"],
]
for group in groups:
    odd = model.doesnt_match(group)
    print(f"  {group} → odd one: '{odd}'")
```

##  GET VECTOR: see the actual numbers behind a word

```python
print("\n" + "=" * 50)
print("5. GET VECTOR for 'hello' (first 10 dimensions)")
print("=" * 50)
vector = model["hello"]
print(f"  Shape: {vector.shape}")        # (300,)
print(f"  First 10 values: {vector[:10].round(4)}")
```

## MOST SIMILAR TO VECTOR: find words near a custom vector

```python
print("\n" + "=" * 50)
print("6. MOST SIMILAR TO VECTOR (average of two words)")
print("=" * 50)
import numpy as np
# Create a vector that is the average of "man" and "woman"
avg_vector = (model["man"] + model["woman"]) / 2
result = model.most_similar(positive=[avg_vector], topn=5)
for word, score in result:
    print(f"  {word:20} → {score:.4f}")
```

## DISTANCE: opposite of similarity (0 = identical)

```python
print("\n" + "=" * 50)
print("7. DISTANCE between words (lower = more similar)")
print("=" * 50)
pairs = [("cat", "dog"), ("cat", "airplane")]
for w1, w2 in pairs:
    dist = model.distance(w1, w2)
    print(f"  {w1:10} ↔ {w2:10} → {dist:.4f}")
```

## SIMILAR BY VECTOR: find similar to positive, away from negative

```python
print("\n" + "=" * 50)
print("8. ANALOGY: Paris is to France as Berlin is to ?")
print("=" * 50)
result = model.most_similar(
    positive=["Berlin", "France"],
    negative=["Paris"],
    topn=3
)
for word, score in result:
    print(f"  {word:20} → {score:.4f}")
```

## VOCABULARY CHECK: is a word in the model?

```python
print("\n" + "=" * 50)
print("9. VOCABULARY CHECK")
print("=" * 50)
words = ["hello", "python", "xyzabc123", "AI"]
for word in words:
    exists = word in model
    print(f"  '{word}': {'✅ found' if exists else '❌ not found'}")
```
